# Ставим код на автоматическое расписание
Довольно часто приходится выполнять скрипты регулярно. Например, обновлять выгрузки из Яндекс.Метрики за прошедший день или неделю, запускать скрипт обработки очередной выгрузки из корпоративной базы данных. Со временем таких скриптов становится все больше. Естественно, надо наладить автоматическое выполнение таких задач.

Способы постановки скриптов на расписание могут сильно зависеть от настроек операционной системы, поэтому мы только познакомимся со стандартными приемами. В некоторых операционных системах есть проблемы с распознаванием адресов файлов с русскими буквами. Поместите свой скрипт в папку только из английских букв. Например, в такую: C:\Users\Username\Scripts

Для постановки скрипта на расписание вам нужно сделать три вещи:

1. Написать команду для запуска вашего скрипта из командной строки

2. Записать эту команду в исполняемый файл (в Windows это файл с расширением bat, в Mac OS и Linux - sh)

3. Назначить вызов этого файла в планировщике задач, который есть в любой операционной системе

Рассмотрим пример такого алгоритма на Windows. Допустим, есть скрипт, который выполняет некоторые действия и в итоге пишет результат в файл data.txt (файл report.ipynb). Важной особенностью постановки задач на расписание является то, что по умолчанию ваши скрипты будут вызываться из вашей домашней директории. Т. е. если в коде есть адрес файла 'data.txt', то система будет искать его в вашей домашней папке (по адресу вида C:\Users\username).

1. Указывать в коде полные адреса всех файлов. Это не очень удобно, т. к. при переносе файла вам придется вручную менять все эти адреса.

2. Воспользоваться библиотекой os. С помощью нее можно узнать папку, в которой сейчас расположен скрипт:
